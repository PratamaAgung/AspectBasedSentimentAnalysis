```python
import requests
from bs4 import BeautifulSoup
import re

headers = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:57.0) Gecko/20100101 Firefox/57.0'}
```

```python
search_keyword = "fashion"
search_url = "https://www.blibli.com/ajax-content?s={}&o=7&i=".format(search_keyword)
r = requests.get(search_url, headers= headers)
data = r.text
```

```python
soup = BeautifulSoup(data, "html.parser")
id_list = []
for detail in soup.find_all("div", class_= "product-detail-wrapper"):
    onclick_func = detail["onclick"] 
    id_list.append(re.findall("'([a-zA-Z0-9\-]*)'", onclick_func)[1])
```

```python
review_api_format = "https://www.blibli.com/backend/product/products/pc--{}/reviews?page={}&itemPerPage={}"

review_list = []
for id in id_list:
    stop = False
    page_no = 1
    page_length = 50
    while (not stop and page_length <= 20):
        print(page_no, id)
        review_api_url = review_api_format.format(id, page_no, page_length)
        data = requests.get(review_api_url, headers= headers).json()
        
        for d in data["data"]["reviews"]:
            review_list.append({
                'id': id,
                'content': d['content'],
                'rating': d['rating']
            })
        
        if len(data["data"]["reviews"]) < page_length:
            stop = True
        else:
            page_no += 1
```

```python
count = 0
for d in review_list:
    if d["rating"] < 3:
        print(d['content'], d['rating'])
        count += 1
count
```

```python
import csv

with open(search_keyword + "_review.csv", "w") as csv_file:
    csv_writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    for review in review_list:
        csv_writer.writerow([review["id"], review['content'], review['rating']])
```
